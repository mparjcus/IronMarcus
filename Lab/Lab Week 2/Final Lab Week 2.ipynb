```python
from sqlalchemy import create_engine
import pandas as pd
import getpass  # To get the password without showing the input
password = getpass.getpass()
```

```python
connection_string = 'mysql+pymysql://root:' + password + '@localhost/sakila'
engine = create_engine(connection_string)
```

```python
query = '''select a.customer_id, first_name, last_name, count(rental_id) as number_of_rentals_may
from rentals_may as a
inner join customer as b
on a.customer_id = b.customer_id
group by a.customer_id
order by count(rental_id) desc'''
may_rentals = pd.read_sql_query(query, engine)
may_rentals.head()
```

```python
may_rentals = pd.DataFrame(may_rentals)
may_rentals.shape
```

```python
query2 = '''select a.customer_id, first_name, last_name, count(rental_id) as number_of_rentals_june
from rentals_june as a
inner join customer as b
on a.customer_id = b.customer_id
group by a.customer_id
order by count(rental_id) desc'''
june_rentals = pd.read_sql_query(query2, engine)
june_rentals.head()
```

```python
june_rentals = pd.DataFrame(june_rentals)
```

```python
june_may_merge = pd.merge(may_rentals, june_rentals, on="customer_id")
june_may_merge.head()
```

```python
june_may_merge = june_may_merge.drop(columns=['first_name_y','last_name_y'])
june_may_merge.head()
```

```python
import numpy as np
```

```python
june_may_merge['Diff'] = np.where(june_may_merge['number_of_rentals_may'] == june_may_merge['number_of_rentals_june'], 0, june_may_merge['number_of_rentals_may'] - june_may_merge['number_of_rentals_june']) #create new column in df1 for price diff 
```

```python
def rental(x,y):
    if x > y:
        return('client lent more movies in may')
    elif y > x:
        return('client lent more movies in june')
    else:
        return('same number')
```

```python
june_may_merge['difference']=list(map(rental, june_may_merge['number_of_rentals_may'], june_may_merge['number_of_rentals_june']))
```
